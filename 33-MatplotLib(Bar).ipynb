```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y Axsis',title='Bar5 With Box')
x=np.linspace(1,21,20)
y=np.random.randint(50,size=20)
plt.bar(x,y,facecolor='black',edgecolor='red',width=0.8,alpha=0.8)
plt.text(7,-10,'Population Density',style='italic',bbox={'facecolor':'gray','alpha':0.5,'pad':10})
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y Axsis',title='Bar 6 With Box And Loctaion Value')
x=np.linspace(1,21,20)
y=np.random.randint(50,size=20)
plt.bar(x,y,facecolor='black',edgecolor='green',width=0.8,alpha=0.8)
plt.text(5,-10,'Population Density',style='italic',bbox={'facecolor':'gray','alpha':0.5,'pad':12})
for xx,yy in zip(x,y):plt.text(xx,yy,yy)#Location Of Text The Value
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Bar 7 With Country Population',xlabel='Country Named',ylabel='Years')
x=np.linspace(1,21,20)
y=np.random.randint(50,size=20)
plt.bar(x,y,facecolor='black',edgecolor='green',width=0.8,alpha=0.5)
plt.text(7,-10,'Population Density',style='italic',bbox={'facecolor':'gray','alpha':0.8,'pad':5})
for xx,yy in zip(x,y):plt.text(xx,yy,yy,ha='center',va='bottom')
```

```python
import matplotlib.pyplot as plt
import numpy as np
num=12
x=np.arange(num)
y1=(1-x/float(num))*np.random.uniform(0.5,1.0,num)
y2=(1-x/float(num))*np.random.uniform(0.5,1.0,num)
plt.bar(x,+y1,color='green',width=0.6,alpha=0.5)
plt.bar(x,-y2,facecolor='orange',width=0.6,alpha=0.5)
plt.text(7,-10,'Two Graphs',style='italic',bbox={'facecolor':'gray','alpha':0.5,'pad':5})
for x,y in zip(x,y1):
    plt.text(x+0.4,y+0.05,'%.2f'%y,ha='center',va='bottom')
    plt.ylim(-1.25,+1.25)
```

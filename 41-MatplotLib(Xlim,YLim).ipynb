```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig,ax=plt.subplots()
ax.set(title='Square Function',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.linspace(-10,10)
data2=-data1**2
plt.plot(data1,data2,linestyle='dashed',marker='*',color='orange',lw=3.5,alpha=0.8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig,ax=plt.subplots()
ax.set(title='Square Function',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.linspace(-10,10)
data2=-data1**2
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
plt.plot(data1,data2,linestyle='dashed',marker='*',color='orange',lw=3.5,alpha=0.8)
```

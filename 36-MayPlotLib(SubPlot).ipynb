```python
import matplotlib.pyplot as plt
import numpy as np
data=np.linspace(0,70,500)
plt.subplot(2,2,1)
data1=data*3
plt.plot(data,data1,color='r',marker='o',alpha=0.5)
plt.subplot(2,2,2)
data2=2*data
plt.plot(data,data2,color='green',marker='^',alpha=0.5)
plt.subplot(2,2,3)
data3=np.sin(data)
plt.plot(data,data3,color='black',alpha=0.5)
plt.subplot(2,2,4)
data4=np.cos(data)
plt.plot(data,data4,color='purple',alpha=0.5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
fig,ax=plt.subplots(2,3,sharex='col',sharey='row')
data=np.linspace(0,50,1000)
plt.subplot(231)
data1=np.sin(data)
plt.plot(data,data1,color='green',linestyle='solid',marker='*',alpha=0.5)
plt.text(-.5,2.5,str((2,3,1,'Sin Function')),fontsize=12,style='italic',ha='left',bbox={'facecolor':'purple','alpha':0.5,'pad':5})
plt.subplot(232)
data2=np.cos(data)
plt.plot(data,data2,color='black',linestyle='dotted',marker='o',alpha=0.5)
plt.text(-.5,2.5,str((2,3,2,'Cos Function')),fontsize=12,style='italic',ha='center',bbox={'facecolor':'purple','alpha':0.5,'pad':5})
plt.subplot(233)
data3=np.exp(data)
plt.plot(data,data3,linestyle='dashed',color='black',marker='x',alpha=0.5)
plt.text(-.5,2.5,str((2,3,3,'Exp Function')),fontsize=12,style='italic',ha='right',bbox={'facecolor':'purple','alpha':0.5,'pad':5})
plt.subplot(234)
data4=np.sqrt(data)
plt.plot(data,data4,linestyle='dashdot',color='blue',marker='*',alpha=0.5)
plt.text(-.5,2.5,str((2,3,4,'Sqrt Function')),fontsize=12,style='italic',ha='left',bbox={'facecolor':'purple','alpha':0.5,'pad':5})
plt.subplot(235)
data5=np.sin(data)+np.pi*data
plt.plot(data,data5,color='yellow',linestyle='dashed',marker='>',alpha=0.5)
plt.text(-.5,2.5,str((2,3,5,'Sin Function')),fontsize=12,style='italic',ha='center',bbox={'facecolor':'purple','alpha':0.5,'pad':5})
plt.subplot(236)
data6=np.tan(data*np.pi)
plt.plot(data,data6,color='purple',linestyle='dotted',marker='<',alpha=0.5)
plt.text(-.5,2.5,str((2,3,6,'Tan Function')),fontsize=12,style='italic',ha='right',bbox={'facecolor':'purple','alpha':0.5,'pad':5})
plt.subplots_adjust(left=.5,right=1.5,bottom=1,top=1.8,wspace=0.5,hspace=0.5)
```

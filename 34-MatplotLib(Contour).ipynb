```python
import matplotlib.pyplot as plt
import numpy as np
def f(x,y):
    var=np.sin(x)**10+np.cos(10+y*x)
    return var
data1=np.linspace(0,5,50)
data2=np.linspace(0,5,40)
X,Y=np.meshgrid(data1,data2)
Z=f(X,Y)
plt.style.use('ggplot')
plt.contour(X,Y,Z,color='gray')
```

```python
#Read Image And Save Allp Pixels In Text File
a=plt.imread('D:\\Images AI\\8.jpg')#Read Image
F=open('D:\\Images AI\\image_8.txt','w+')#Create File And Start Write In This
var1,var2,var3=a.shape#Three Variables Equal Matrix Value
for x in range(var1):
    for y in range(var2):
        for z in range(var3):
            F.write('\n'+str('Date For Image Is => '+str(x)+' and '+str(y) + ' and ' + str(z) +' is ==> '+str(a[x,y,z])))
F.close()
plt.imshow(a)
```

```python
import matplotlib.patches as pat
import matplotlib.pyplot as plt
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Circle Shape',xlabel='X Axsis',ylabel='Y Axsis')
c=pat.Circle((0.5,0.5),radius=0.2)
ax.add_patch(c)
```

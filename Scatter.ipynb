```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y Axsis')
ax.set(title='Scatter1')
x=np.random.randn(50)
y=np.random.randn(50)
plt.scatter(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax=plt.axes()
plt.style.use('ggplot')
ax.set(xlabel='X Axsis',ylabel='Y Axsis')
ax.set(title='Scatter Normal Distrbution')
x=np.random.normal(0,1,100)
y=np.random.normal(0,1,100)
plt.scatter(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y Axsis')
ax.set(title='Scatter 3 With Drwing Markers')
x=np.random.normal(0,1,200)
y=np.random.normal(0,1,200)
z=np.random.normal(0,1,200)
plt.scatter(x,y,c=z)
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax=plt.axes()
plt.style.use('ggplot')
ax.set(xlabel='X Axsis',ylabel='Y Axsis')
ax.set(title='Scatter 3 With Colors And Markers Size')
x=np.random.normal(0,1,500)
y=np.random.normal(0,1,500)
z=np.random.normal(0,1,500)
w=30*np.random.normal(0,1,500)
plt.scatter(x,y,s=w,c=z)
```

```python
import matplotlib.pyplot as plt
import numpy as np
rng=np.random.RandomState(0)
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y Acsis')
ax.set_title('Scatter 4 With Alpha')
x=rng.randn(100)
y=rng.randn(100)
cl=rng.rand(100)
sz=1000*rng.rand(100)
plt.scatter(x,y,s=sz,c=cl,alpha=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
rng=np.random.RandomState(0)
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y Axsis')
ax.set_title('Scatter 4 With ColorBar')
x=rng.randn(100)
y=rng.randn(100)
sz=1000*rng.rand(100)
cl=rng.rand(100)
plt.scatter(x,y,s=sz,c=cl,alpha=0.3)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
fig,ax=plt.subplots()
ax.set(title='Legend Shape_1',xlabel='X Axsis',ylabel='Y Axsis')
data=np.linspace(0,2,1000)
ax.plot(data,np.sin(data),color='green',linestyle='dashed',marker='>',alpha=0.5,label='Sin')
ax.plot(data,np.cos(data),color='red',linestyle='dotted',marker='*',markersize=2.5,alpha=0.5,label='Cos')
leg=ax.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
fig,ax=plt.subplots()
data=np.linspace(0,10,50)
ax.set(title='Legend Shape_2',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.power(data,3.5)*np.exp(-data/2)
data2=np.cos(data)*np.power(data,2)
data3=np.power(data,2.5)
data4=np.sin(data)*np.cos(data)*np.tan(data)+np.sqrt(data)
plt.plot(data,data1,linestyle='dashed',lw=1,color='green',alpha=0.5,marker='*',label='Sin')
plt.plot(data,data2,linestyle='dashed',color='red',lw=1,alpha=0.5,marker='*',label='Cos')
plt.plot(data,data3,linestyle='solid',color='orange',lw=1,marker='>',alpha=0.5,label='Exp')
plt.plot(data,data4,linestyle='dotted',color='blue',lw=1,alpha=0.5,marker='^',label='Tan')
leg=ax.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
fig,ax=plt.subplots()
data=np.linspace(0,10,50)
ax.set(title='Legend Shape_2',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.power(data,3.5)*np.exp(-data/2)
data2=np.cos(data)*np.power(data,2)
data3=np.power(data,2.5)
data4=np.sin(data)*np.cos(data)*np.tan(data)+np.sqrt(data)
data5=np.pi*data
data6=np.power(data,1.3)*np.sin(data)+np.cos(data)*np.tan(data)
plt.plot(data,data1,linestyle='dashed',lw=1,color='green',alpha=0.5,marker='*',label='Sin')
plt.plot(data,data2,linestyle='dashed',color='red',lw=1,alpha=0.5,marker='*',label='Cos')
plt.plot(data,data3,linestyle='solid',color='orange',lw=1,marker='>',alpha=0.5,label='Exp')
plt.plot(data,data4,linestyle='dotted',color='blue',lw=1,alpha=0.5,marker='^',label='Tan')
plt.plot(data,data5,color='purple',linestyle='dotted',lw=1,alpha=0.5,label='Pi')
plt.plot(data,data6,linestyle='dashed',lw=1,color='black',alpha=0.5,label='Complex')
leg=ax.legend()
ax.legend(loc='upper left',frameon=True,ncol=3,fancybox=True,shadow=True,framealpha=0,borderpad=1,fontsize=12)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
fig,ax=plt.subplots()
data=np.linspace(0,20,1000)
ax.set(title='Complex Shape',xlabel='X Axsis',ylabel='Y Axsis')
data6=np.power(data,1.3)*np.sin(data)+np.cos(data)*np.tan(data)
Lines=plt.plot(data,data6,linestyle='dashed',lw=1,color='black',alpha=0.5)
plt.legend(Lines[:2],['Cost','Time'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 20, 1000)
fig, ax = plt.subplots()
plt.axis('equal')
y = np.sin(x[:, np.newaxis] + np.pi *
np.arange(0, 2, 0.5))
lines = plt.plot(x, y)
# lines is a list of plt.Line2D instances
plt.legend(lines[:2], ['Cost', 'Time'])
```

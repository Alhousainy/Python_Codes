```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
fig,ax=plt.subplots()
data1=np.linspace(0,20,1000)
data2 = -((data1-10)**2) +5
ax.set(title='Anntotate Shape_1',xlabel='X Axsis',ylabel='Y Axsis')
plt.plot(data1,data2,linestyle='dotted',color='green',marker='>',alpha=0.5)
plt.annotate('max',xy=(10,6),xytext=(15,10),arrowprops=dict(color='purple',shrink=0.2,lw=1))
plt.annotate('min',xy=(5,-20),xytext=(5,6),arrowprops=dict(color='red',shrink=0.2,lw=1))
plt.annotate('value',xy=(1.5,-85),xytext=(4,-85),arrowprops=dict(color='gray',shrink=0.2,lw=1))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()
plt.style.use('ggplot')
ax.set(title='Annotate Curvey Arrow',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.linspace(0,20,1000)
data2=np.cos(data1)
ax.plot(data1,data2,color='red',linestyle='dotted',marker='^',alpha=0.5)
ax.axis('equal')
plt.text(8.5,-11.2,'Arrow Curvey',style='italic',fontsize=12,color='green',bbox={'facecolor':'gray','alpha':0.5,'pad':5})
plt.annotate('Local Maximum',xy=(6.28,1),xytext=(10,4),arrowprops=dict(color='black',shrink=1.5,lw=2))
plt.annotate('local Minimum',xy=(5*np.pi,-1),xytext=(2,-6),arrowprops=dict(color='blue',shrink=1.5,lw=1.2,connectionstyle='angle3,angleA=0,angleB=-90'))
plt.annotate('Theta',xy=(7.5,0),xytext=(1.5,-4),arrowprops=dict(color='orange',shrink=1.5,lw=2,connectionstyle='angle3,angleA=-15,angleB=-60'))
plt.annotate('theat2',xy=(10.5,-.98),xytext=(10,-4),arrowprops=dict(color='green',shrink=1.5,lw=3,connectionstyle='angle3,angleA=-20,angleB=-50'))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
fig,ax=plt.subplots()
data1=np.linspace(0,50,1000)
data2=np.exp(-data1/2)*np.cos(data1)
ax.set(title='Curvey_ Arrow_Shape_2',xlabel='X Axsis',ylabel='Y Axsis')
plt.plot(data1,data2,color='orange',alpha=0.5)
plt.text(21,-.6,'Curvery_arrow',style='italic',fontsize=12,color='green',bbox={'facecolor':'gray','alpha':0.5,'pad':5})
plt.annotate('Local Maximum',xy=(20,0),xytext=(25,0.2),arrowprops=dict(color='black',shrink=1.2,lw=1))
plt.annotate('value_1',xy=(10,0),xytext=(5,.3),arrowprops=dict(color='blue',shrink=1.7,lw=1.2))
plt.annotate('local_Minimum',xy=(15,0),xytext=(5,-0.2),arrowprops=dict(color='green',connectionstyle='angle3,angleA=0,angleB=-35'))
plt.annotate('Value_2',xy=(40,0),xytext=(45,-0.25),arrowprops=dict(color='red',connectionstyle='angle3,angleA=0,angleB=90'))
```

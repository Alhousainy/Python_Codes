```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Histogram 2 Graphs Together',xlabel='X Axsis',ylabel='Y Axsis')
data=np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data1=pd.DataFrame(data,columns=['x','y'])
for col in 'xy':plt.hist(data1[col],alpha=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Historam 3 Graphs Together',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.random.normal(0,1.5,500)
data2=np.random.normal(1,3.5,500)
data3=np.random.normal(2,4.3,500)
plt.hist(data1,bins=100,histtype='step',alpha=0.8)
plt.hist(data2,bins=100,histtype='bar',alpha=0.8)
plt.hist(data3,bins=100,histtype='stepfilled',alpha=0.8)
plt.text(-4,-5,'Three Graphs',style='italic',bbox={'facecolor':'gray','alpha':0.5,'pad':8})
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Historam 3 Graphs Together',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.random.normal(0,1.5,500)
data2=np.random.normal(1,3.5,500)
data3=np.random.normal(2,4.3,500)
kwarg={'histtype':'stepfilled','alpha':0.8,'bins':50}
plt.text(-4,-5,'Three Graphs',style='italic',bbox={'facecolor':'gray','alpha':1,'pad':5})
plt.hist(data1,**kwarg)
plt.hist(data2,**kwarg)
plt.hist(data3,**kwarg)
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax=plt.axes()
plt.style.use('ggplot')
ax.set(title='Histogram Different Kwargs Method ',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.random.normal(0,2,500)
data2=np.random.normal(1,3,500)
data3=np.random.normal(3,5,500)
kwarg1={'histtype':'bar','alpha':0.5,'color':'gray'}
kwarg2=dict(histtype='stepfilled',alpha=0.5,color='orange')
kwarg3={'histtype':'step','alpha':0.5,'color':'green'}
plt.text(0,-25,'Three Graphs',style='italic',bbox={'facecolor':'blue','alpha':0.5,'pad':4})
plt.hist(data1,**kwarg1)
plt.hist(data2,**kwarg2)
plt.hist(data3,**kwarg3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
D=plt.axes()
D.set(title='Normal Distrbution',xlabel='X Axsis',ylabel='Y Axsis')
mean=[0,0]
cov=[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,1000).T
plt.hist2d(x,y,cmap='Blues',bins=50)
ax=plt.colorbar()
ax.set_label('Normal Distrbution Effects')
```

```python
import matplotlib.pyplot as plt
import numpy as np
mean=[0,0]
cov=[[1,1],[1,2]]
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Normal Distrbution using hexbin',xlabel='X Axsis',ylabel='Y Axsis')
data1,data2=np.random.multivariate_normal(mean,cov,1000).T
plt.hexbin(data1,data2,cmap='Blues',gridsize=30,bins=50)
D=plt.colorbar()
D.set_label('Counts In Bins')
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
fig, ax = plt.subplots()
data1 = np.linspace(0, 20, 1000)
data2 = -((data1-10)**2) +5
ax.plot(data1,data2,color='black',alpha=0.5,linestyle='dashed',marker='*')
plt.annotate('Max',xy=(10,6),xytext =(15,10),arrowprops =dict(color='r',shrink=0.02,lw =1.5))
plt.annotate('5',xy=(5,-20),xytext =(2,0),arrowprops =dict(color='b',shrink=3,lw =5))
plt.annotate('15',xy=(15,-20),xytext =(10,-60),arrowprops =dict(color='g',shrink=3,lw =9))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
fig,ax=plt.subplots()
data1=np.linspace(0,20,1000)
ax.set(title='Annotate Graph1',xlabel='X Axsis',ylabel='Y Axsis')
data2= -((data1-10)**2) +5
plt.text(8,-120,'Anntotate_1',style='italic',fontsize=12,color='red',bbox={'facecolor':'gray','alpha':0.5,'pad':5})
plt.plot(data1,data2,color='black',linestyle='dotted',alpha=0.5)
plt.annotate('max',xy=(10,6),xytext=(15,10),arrowprops=dict(color='orange',shrink=0.03,lw=1.2))
plt.annotate('value_2',xy=(5,-20),xytext=(2,0),arrowprops=dict(color='blue',shrink=1.5,lw=1.5))
plt.annotate('value_3',xy=(15,-20),xytext=(10,-60),arrowprops=dict(color='red',shrink=2,lw=2))
plt.annotate('value_4',xy=(2.5,-50),xytext=(5,-50),arrowprops=dict(color='green',shrink=3,lw=1.8))
plt.annotate('value_5',xy=(7.5,0),xytext=(7.5,-20),arrowprops=dict(color='black',shrink=2,lw=1.7))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()
plt.style.use('ggplot')
ax.set(title='Annotate Curvey Arrow',xlabel='X Axsis',ylabel='Y Axsis')
data1=np.linspace(0,20,1000)
data2=np.cos(data1)
ax.plot(data1,data2,color='red',linestyle='dotted',marker='^',alpha=0.5)
ax.axis('equal')
plt.text(8.5,-11.2,'Arrow Curvey',style='italic',fontsize=12,color='green',bbox={'facecolor':'gray','alpha':0.5,'pad':5})
plt.annotate('Local Maximum',xy=(6.28,1),xytext=(10,4),arrowprops=dict(color='black',shrink=1.5,lw=2))
plt.annotate('local Minimum',xy=(5*np.pi,-1),xytext=(2,-6),arrowprops=dict(color='blue',shrink=1.5,lw=1.2,connectionstyle='angle3,angleA=0,angleB=-90'))
plt.annotate('Theta',xy=(7.5,0),xytext=(1.5,-4),arrowprops=dict(color='orange',shrink=1.5,lw=2,connectionstyle='angle3,angleA=-15,angleB=-60'))
plt.annotate('theat2',xy=(10.5,-.98),xytext=(10,-4),arrowprops=dict(color='green',shrink=1.5,lw=3,connectionstyle='angle3,angleA=-20,angleB=-50'))
```

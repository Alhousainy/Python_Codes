```python
import matplotlib.pyplot as plt
import numpy as np
a=(1,2,3,4,5,6)
b=(3,9,10,12,16,18)
plt.style.use('ggplot')
plt.ylabel('Time')
plt.xlabel('Work')
plt.plot(a,b)
```

```python
import matplotlib.pyplot as plt
a=(1,2,3,4,5,6)
b=(3,9,10,12,16,18)
ax=plt.axes()
ax.set_xlabel('Time')
ax.set_ylabel('Work')
plt.plot(a,b)
```

```python
import matplotlib.pyplot as plt
a=(1,2,3,4,5,6)
b=(3,9,10,12,16,18)
ax=plt.axes()
ax.set(xlabel='Time',ylabel='Work')
plt.plot(a,b)
```

```python
import matplotlib.pyplot as plt
a=(1,2,3,4,5,6)
b=(3,9,10,12,16,18)
ax=plt.axes()
ax.set(xlabel='Work',ylabel='Time')
plt.plot(a,b,marker='o')
```

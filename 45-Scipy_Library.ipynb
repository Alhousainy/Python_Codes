```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.optimize import curve_fit
plt.style.use('ggplot')
ax=plt.axes()
ax.set(title='Best Fit Line ',xlabel='X Axsis',ylabel='Y Axsis')
#Fucntion To Calculate Value
def func(x,a,b,c):return a*np.exp(-b*x)+c
#Random Values Of X Data
xdata=np.linspace(0,4,50)
y=func(xdata,2.5,1.3,0.5)
#Y Is To Call Function And Send Xdata to X in Function & a_value =2.5 ,b_value=1.5 & c_value  =0.5
np.random.seed(1729)#Random.seed ==> Is a numerical value that generates a new set or repeats pseudo-random numbers . the value in the numpy random seed saves the state of randomness.
#Y_Noise ==> I need to add positive and negative random values to xdata untill make a fitting
y_noise=0.2*np.random.normal(size=xdata.size)
ydata=y+y_noise
plt.plot(xdata,ydata,color='blue',linestyle='dashed',label='data')#Draw The Line That Represent xdata and the added values(ydata)
#==============================================================================================================================
popt,pcov=curve_fit(func,xdata,ydata)#Use Curve Fit To Make Fitting for The Best Fit Line 
#popt ==> is equal func & pcov is ==> xdata, ydata
plt.plot(xdata,func(xdata,*popt),'r-'#Draw The Line 
,label='Fit : a=%5.3f , b = %5.3f , c = %5.3f'%tuple(popt))
#===============================================================================================================================
#Fitting With Boundaries 
popt,pcov=curve_fit(func,xdata,ydata,bounds=(0,[3.,1.,0.5]))
plt.plot(xdata,func(xdata,*popt),#Draw the green line that represent boundaries line
'g--',label='Fit Is : a = %5.3f , b = %5.3f , c = %5.3f ' %tuple(popt))
#================================================================================================================================
plt.legend()
```
